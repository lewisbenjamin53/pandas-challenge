```python
# Dependencies and Setup
import pandas as pd


# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
df = pd.read_csv(file_to_load)
df.head()
```

```python
#create bins and their group names
bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 100]
groups = ["<10", "10-14", "15-19", "20-24","25-29", "30-34", "35-39", "40+" ]

#creating and slicing dataframe & add new column called "Age group"
age_bin_df = df
age_bin_df["Age Group"] = pd.cut(df["Age"], bins, labels = groups)

#using groupby based on age group
age_bin_group = age_bin_df.groupby("Age Group").count()

#Create new columns for Total Count and Percentage
age_bin_group["Total Count"] = age_bin_group["SN"]
age_bin_group["Percentage of Players"] = 100*(age_bin_group["Total Count"]/(age_bin_group["Total Count"].sum()))

age_demo_df = age_bin_group[["Total Count","Percentage of Players"]]
age_demo_df
```

```python
#adding columns to this dataframe
age_analysis = age_demo_df[["Total Count"]]
age_analysis = age_analysis.rename(columns={"Total Count" : "Purchase Count"})

#calculating other columns
age_bin_avg = age_bin_df.groupby("Age Group").mean()
age_analysis[["Average Purchase Price"]] = age_bin_avg[["Price"]]
age_bin_sum = age_bin_df.groupby("Age Group").sum()
age_analysis[["Total Purchase Value"]] = age_bin_sum[["Price"]]


#print out age analysis dataframe
age_analysis
```

```python
user_count = df.groupby("SN").count()

top_spend = user_count[["Price"]]
top_spend = top_spend.rename(columns = {"Price" : "Purchase Count"})


user_avg = df.groupby("SN").mean()
top_spend["Average Purchase Price"] = user_avg["Price"]

user_sum = df.groupby("SN").sum()
top_spend["Total Purchase Value"] = user_sum["Price"]

top_spend.sort_values("Purchase Count", ascending = False).head()
```

```python
#creating dataframe
gb_df = df.groupby(["Item ID", "Item Name"])

#calculations
item_count = gb_df["Price"].count()
item_avg = gb_df["Price"].mean()
item_tot = gb_df["Price"].sum()

top_spend_sum = pd.DataFrame([item_avg, item_count, item_tot])
top_spend_sum_t = pd.DataFrame(top_spend_sum.T)
top_spend_sum_t.columns = ["Purchase Count", "Item Price", "Total Purchase Value"]

top_spend_sum_t.head()

top_spend_sum_t.sort_values("Purchase Count", ascending = False).head()
```
